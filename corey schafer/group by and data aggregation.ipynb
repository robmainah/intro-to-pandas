```python
country_respondents = df['Country'].value_counts()
country_respondents
```

```python
python_coders = countries['LanguageWorkedWith'].apply(lambda x: x.str.contains('Python').sum())
python_coders
```

```python
python_df = pd.concat([country_respondents, python_coders], axis='columns', sort=False)
python_df
```

```python
python_df.rename(columns={'Country': 'NumRespondents', 'LanguageWorkedWith': 'NumKnowsPython'}, inplace=True)
python_df
```

```python
python_df['KnowsPython%'] = (python_df['NumKnowsPython']/python_df['NumRespondents']) * 100
python_df
```

```python
python_df.sort_values(['KnowsPython%'], ascending=False, inplace=True)
python_df
```
